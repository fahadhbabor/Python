```python
def avg_calc(units):
    slabs = [
        (50, 4.63),
        (25, 5.26),
        (125, 7.20),
        (100, 7.59),
        (100, 8.02),
        (200, 12.69),
        (float('inf'), 14.61),
    ]

    total_energy_cost = 0
    remaining = units

    for slab_units, rate in slabs:
        if remaining <= 0:
            break
        used = min(remaining, slab_units)
        total_energy_cost += used * rate
        remaining -= used

    # VAT 4.8%
    vat = total_energy_cost * 0.048

    # Final bill
    total_cost = total_energy_cost + vat

    avg_price = total_cost / units

    return avg_price
```

```python
# Example usage
units = 100

avg = avg_calc(units)

print(f"Avg Rate:    {avg:.4f}/unit")
```

```python
# Example usage
units = 213.46

avg = avg_calc(units)

print(f"Avg Rate:    {avg:.4f}/unit")
```
