# Global variable

```python
x = "awesome"

def myfunc():
    print("Python is " + x)

myfunc()
```

```python
x = "awesome" #Global variable

def myfunc():
  #Local variable
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

# global keyword

```python
def myfunc():
    global x
    x = "fantastic"

myfunc()
print("Python is "+ x)
```
