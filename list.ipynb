# List 

List is a collection of items in a particular order.

```python
bicycles = ['trek','cannondale','redline','specialized']
print(bicycles)
```

```python
print(bicycles[0]) #prints the first element && indexing start from 0
```

```python
bicycles[0].title() #Title case 
```

```python
print(bicycles[3])
```

```python
#negative indexing 
print(bicycles[-1])
```

```python
message = f"My first bicycle was a {bicycles[0].title()}"
print(message)
```

## Removing

```python
names = ['setu','muntaha','aysha','aru']
del names[-1] #removing elements via 'del' keyword
print(names)
```

```python
names.insert(0,'Jannat')
```

```python
print(names)
```

### Removing elements using pop() method

```python
last_name = names.pop()
print(names)
print(last_name)
```

```python
names = ['setu','muntaha','aysha','aru']
print(f"My last gf was {names.pop().title()}.")
```

### popping from any position

```python
first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}")
```

### removing an item via value 

```python
names = ['setu','muntaha','aysha','aru']
names.remove('aysha')
print(names)
```

# Organizing list 

```python
#sort function is permanent 
cars = ['bmw','audi','honda','toyota','subaru']
cars.sort()
print(cars)
```

```python
cars.sort(reverse=True) # reverse the list
```

```python
print(cars)
```

```python
## Temporary sort- doesnt change the original list 
cars = ['bmw','audi','honda','toyota','subaru']
print(sorted(cars))
print(cars)
```

```python
#does permanent damage - just reverse the list
cars.reverse()
print(cars)
```

```python
sorted(cars,reverse=True) # sort it in reverse order 
```
